```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import threading
from IPython.display import clear_output


def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        soup = BeautifulSoup(html.read(), "lxml")
    except AttributeError as e:
        return None
    return soup

def start():
    clear_output()
    url = "http://www.reuters.com/"
    soup = getTitle(url)

    print ("Latest news which is keep updating")
    for text3 in soup.find_all('article', attrs={'class':re.compile('^article')} ):
        time = text3.find('span', attrs={'class':'timestamp'}).get_text()
        for text4 in text3.find_all('a'):
            href = text4.attrs["href"]
            title = text4.get_text().strip()
            print('Title-', title)
            print('Time:', time)
            print('website: http://www.reuters.com'+href)
    print ("Main News")
    for text in soup.find_all('div', attrs={'class':'story-content'} ):
        for text2 in text.find_all('a'):
            href = text2.attrs["href"]
            title = text2.get_text().strip()
            print('Title-', title)
            print('website: http://www.reuters.com'+href)
            
def setInterval(func,time):
    e = threading.Event()
    while not e.wait(time):
        func()

setInterval(start,60)
```
